```python
def male_female_child (passenger):
    age,sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
sns.catplot(x="Pclass", data=titanic_df, hue= 'person',kind="count", palette="mako")
plt.title("Podzia wedug osb w klasach pasaerskich")
plt.show()
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])

cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
cabin_df = cabin_df.sort_values(by='Cabin')
sns.catplot(data=cabin_df, x ='Cabin', hue= 'Cabin',kind="count",palette= 'mako', legend = False)
plt.show()
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
sns.catplot(data=cabin_df, x ='Cabin', hue= 'Cabin',kind="count",palette= 'summer', legend = False)
plt.show()
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
titanic_df['Alone']
```

```python
titanic_df['Alone'].loc[titanic_df['Alone']>0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] ==0] = 'Alone'
```

```python
titanic_df.head()
```

```python
sns.catplot(data=titanic_df, x ='Alone', hue= 'Alone',kind="count",palette= 'Blues', legend = False)
plt.show()
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0: 'no', 1: 'yes'})
```

```python
sns.catplot(data=titanic_df, x ='Survivor', hue= 'Survivor',kind="count",palette= 'Set1', legend = False)
plt.show()
```

```python
sns.catplot(data=titanic_df, x='Pclass', y= 'Survived', kind = 'point')
plt.show()
```

```python
sns.catplot(data=titanic_df, x='Pclass', y= 'Survived',hue= 'person', kind = 'point')
plt.show()
```

```python
sns.lmplot(x ='Age',y ='Survived', data=titanic_df)
plt.show()
```

```python
generations = [10,20,40,60,80]
sns.lmplot(x ='Age',y ='Survived', hue='Pclass', data=titanic_df, palette = 'winter', x_bins=generations)
plt.show()
```

```python
sns.lmplot(x ='Age',y ='Survived', hue='Sex', data=titanic_df, palette = 'winter', x_bins=generations)
plt.show()
```
